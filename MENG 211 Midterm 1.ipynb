```python
#Problem 3

import numpy as np
import matplotlib.pyplot as plt

#Part A

x_array = np.array(np.linspace(-5, 5, 80))

def a_func(x):
    return x**2 - 4*x + 3

y_val = a_func(x_array)

plt.figure(figsize=(10, 5), dpi=120)
plt.subplot(111)
plt.plot(x_array, y_val, color="purple", marker = 'o',linestyle = 'none', label = 'function of x')
#plt.semilogy(array[3], array[2], color="purple", linewidth=2.5, linestyle="-", label = 'True Error')
#plt.fill_between(array[3], array[2], 0, alpha=0.2, color = 'b') 

plt.title('Graph of x^2 - 4x + 3', fontsize = 18)    
plt.xlabel('value of x', fontsize = 12)
plt.ylabel('function output', fontsize = 12)
plt.legend(loc = 'best')
plt.show()
```

```python
#Part B

def g_one(g):
    return g**2 - 3*g +3

def g_two(g):
    return -(3/(g-4))

def g_three(g):
    return np.sqrt(4*g - 3)
```

```python
#Part C

def f_p_iter(f,x,n):
    for i in range(0,n):
        x = f(x)
    return x

print(f_p_iter(g_one,1.3,15))
print(f_p_iter(g_two,1.3,15))
print(f_p_iter(g_three,1.3,15))
```

```python
#Part D

plt.figure(figsize=(10, 5), dpi=120)
plt.subplot(111)

for n in range(1,16):
    plt.plot(n, f_p_iter(g_one,1.3,n), color="purple", marker = 'o',linewidth=2.5, linestyle="none", label=f"True Error for x=g_one" if n == 1 else "")
    plt.plot(n, f_p_iter(g_two,1.3,n), color="r",  marker = 'o',linewidth=2.5, linestyle="none", label=f"True Error for x=g_two" if n == 1 else "")
    plt.plot(n, f_p_iter(g_three,1.3,n), color="b",  marker = 'o',linewidth=2.5, linestyle="none", label=f"True Error for x=g_three" if n == 1 else "")

plt.title('x vs nth iteration', fontsize = 18)    
plt.xlabel('value of n', fontsize = 12)
plt.ylabel('value of x', fontsize = 12)
plt.legend(loc = 'best')
```

```python
#Part E

def first_derivative_central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

x_array = np.array(np.linspace(0.8, 3.5, 30))

y_one = []
y_two = []
y_three = []

for i in range(0, len(x_array)):
    y_one.append(first_derivative_central(g_one, x_array[i], 0.05))
    y_two.append(first_derivative_central(g_two, x_array[i], 0.05))
    y_three.append(first_derivative_central(g_three, x_array[i], 0.05))

plt.figure(figsize=(10, 5), dpi=120)
plt.subplot(111)
plt.plot(x_array, y_one, color="purple", marker = 'none',linestyle = '-', label = 'g_one')
plt.plot(x_array, y_two, color="b", marker = 'none',linestyle = '-', label = 'g_two')
plt.plot(x_array, y_three, color="r", marker = 'none',linestyle = '-', label = 'g_three')
plt.title('derivitives of g1(x), g2(x), g3(x)', fontsize = 18)    
plt.xlabel('value of x', fontsize = 12)
plt.ylabel('derivitive value', fontsize = 12)
plt.legend(loc = 'best')
plt.show()
```

Part F \

15 iterations, the function was still not converged to a value of x \
for g_one. It did appear to converge to two different values for both \
g2 and g3. Looking at the derivitive graph and the values in part d, \

x=2.5. I think a major problem here is that these are approaching the root \
of the derivitives very slowly and getting stuck where the derivitive is \
equal to 0, rather than solving for the proper root of the given function.
