```python
#Bisect Method

def bisect_method(function, a, b, n):
    for i in range(0,n):
        x = (1/2)*(a+b)
        guess = function(x)
        value = guess*function(b)
        if value > 0:
            b = x
        elif value < 0:
            a = x
            a = x
        else:
            return [x,i]
    return x

def bisect_method_two(function, a, b, error):
    x = (1/2)*(b+a)
    guess = function(x)
    n=1
    while np.abs(guess) > error:
        guess = function(x)
        value = guess*function(b)
        if value > 0:
            b = x
        elif value < 0:
            a = x
        x = (1/2)*(a+b)
        n += 1
        if n > 150:
            return [x, n]
    return [x, n] # [x value, number of iterations]
```

```python
#False Position Method
def FalsePosition(func, lower, upper, n):
    i = 0
    FalsePosition_every_iter = []  # stores a list of h values for every iteration of depth_solve_fp
    
    while i < n:
        xr = upper - (func(upper)*(lower-upper)) / (func(lower)-func(upper))
        test = func(lower)*func(xr)
        
        if test < 0:
            upper = xr
        elif test > 0:
            lower = xr

        i += 1
        FalsePosition_every_iter.append(xr)
        
    return xr, FalsePosition_every_iter  # returns final value of depth_solve_bisect and false_position list
```

```python
#Newton Raphson

def newton_raphson(function, derivitive, guess, n): #This is for a specific amount of iterations
    guess_list = []
    for i in range(0,n):
        guess_list.append(guess)
        guess = guess - function(guess)/derivitive(guess)
    return [guess, guess_list]
```

```python
#From Desktop
def central_first_derivative(f, x, h):
    '''
    Approximates the first derivative of f at x using the central difference method.
    
    Parameters:
        f (function): The function for which to compute the derivative.
        x (float): The point at which to evaluate the derivative.
        h (float): The step size for finite differences. Should be a small positive number.
        
    Returns:
        float: The approximate first derivative of f at x.
        
    Order of Error:
        O(h^2) - Second order accurate
    '''
    return (f(x + h) - f(x - h)) / (2 * h)

def forward_first_derivative(f, x, h):
    '''
    Approximates the first derivative of f at x using the forward difference method.
    
    Parameters:
        f (function): The function for which to compute the derivative.
        x (float): The point at which to evaluate the derivative.
        h (float): The step size for finite differences. Should be a small positive number.
        
    Returns:
        float: The approximate first derivative of f at x.
        
    Order of Error:
        O(h) - First order accurate
    '''
    return (f(x + h) - f(x)) / h

def backward_first_derivative(f, x, h):
    '''
    Approximates the first derivative of f at x using the backward difference method.
    
    Parameters:
        f (function): The function for which to compute the derivative.
        x (float): The point at which to evaluate the derivative.
        h (float): The step size for finite differences. Should be a small positive number.
        
    Returns:
        float: The approximate first derivative of f at x.
        
    Order of Error:
        O(h) - First order accurate
    '''
    return (f(x) - f(x - h)) / h

```

```python
def first_derivative_central(f, x, h):
    """Compute the first derivative at x using central difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the first derivative at x.
    """
    return (f(x + h) - f(x - h)) / (2 * h)

def first_derivative_forward(f, x, h):
    """Compute the first derivative at x using forward difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the first derivative at x.
    """
    return (f(x + h) - f(x)) / h

def first_derivative_backward(f, x, h):
    """Compute the first derivative at x using backward difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the first derivative at x.
    """
    return (f(x) - f(x - h)) / h

def first_derivative_central_higher(f, x, h):
    """Compute the first derivative at x using a higher-order central difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the first derivative at x with higher accuracy.
    """
    return (f(x + 2 * h) - 4 * f(x) + 6 * f(x - h) - 4 * f(x - 2 * h)) / (12 * h)

def second_derivative_central(f, x, h):
    """Compute the second derivative at x using central difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the second derivative at x.
    """
    return (f(x + h) - 2 * f(x) + f(x - h)) / (2 * h)

def third_derivative_central(f, x, h):
    """Compute the third derivative at x using central difference method.
    
    Args:
        f: The function to compute the derivative of.
        x: The point where the derivative is computed.
        h: The step size.
        
    Returns:
        float: The value of the third derivative at x.
    """
    return (f(x + h) - 3 * f(x) + 3 * f(x - h) - f(x - 2 * h)) / (2 * h)

def example_usage():
    import numpy as np
    f = lambda x: np.cos(x)
    x = 0.0
    h = 0.1
    
    # Compute first derivative using different methods
    fd_central = first_derivative_central(f, x, h)
    fd_forward = first_derivative_forward(f, x, h)
    fd_backward = first_derivative_backward(f, x, h)
    fd_central_higher = first_derivative_central_higher(f, x, h)
    
    print("First Derivative at:", x)
    print("Central Difference: ", fd_central)
    print("Forward Difference: ", fd_forward)
    print("Backward Difference: ", fd_backward)
    print("Higher-order Central: ", fd_central_higher)
    
    # Compute second derivative
    sd_central = second_derivative_central(f, x, h)
    
    print("\nSecond Derivative at:", x)
    print("Central Difference: ", sd_central)
    
    # Compute third derivative
    td_central = third_derivative_central(f, x, h)
    #td_forward = third_derivative_forward(f, x, h)
    #td_backward = third_derivative_backward(f, x, h)
    
    print("\nThird Derivative at:", x)
    print("Central Difference: ", td_central)
    #print("Forward Difference: ", td_forward)
    #print("Backward Difference: ", td_backward)

if __name__ == "__main__":
    example_usage()
```

```python
import numpy as np

def central_fd(y, x, h, order=2):
    """
    Central finite difference method for a given order.
    
    Args:
        y (list or array): Array of values.
        x (int or float): Point around which to compute the derivative.
        h (float): Step size.
        order (int): Order of the derivative (1, 2, 3).
        
    Returns:
        float: Approximation of the derivative at point x.
    """
    n = len(y)
    if n < order + 2 * 2:
        raise ValueError("Not enough points to compute the derivative.")
    
    if order == 1:
        return np.round((y[x+h] - y[x-h]) / (2*h), 6)
    elif order == 2:
        return np.round((y[x+h] - 2*y[x] + y[x-h]) / h**2, 6)
    elif order == 3:
        if len(y) >= 4:
            return np.round((y[x+2*h] - 3*y[x+h] + 3*y[x] - y[x-2*h]) / (2*h**3), 6)
        else:
            raise ValueError("Not enough points to compute third derivative.")
    else:
        raise ValueError("Order must be 1, 2, or 3.")

def forward_fd(y, x, h, order=2):
    """
    Forward finite difference method for a given order.
    
    Args:
        y (list or array): Array of values.
        x (int or float): Point around which to compute the derivative.
        h (float): Step size.
        order (int): Order of the derivative (1, 2, 3).
        
    Returns:
        float: Approximation of the derivative at point x.
    """
    n = len(y)
    if n < order + 2 * 2:
        raise ValueError("Not enough points to compute the derivative.")
    
    if order == 1:
        return np.round((y[x+h] - y[x]) / h, 6)
    elif order == 2:
        if len(y) >= 3:
            return np.round(((y[x+2*h] - y[x+h])/(h*2) - (y[x+h] - y[x])/h) / 2, 6)
        else:
            raise ValueError("Not enough points to compute second derivative.")
    elif order == 3:
        if len(y) >= 4:
            return np.round(((y[x+3*h] - 3*y[x+2*h] + 3*y[x+h] - y[x])/(h**3) * (h/2)), 6)
        else:
            raise ValueError("Not enough points to compute third derivative.")
    else:
        raise ValueError("Order must be 1, 2, or 3.")

def backward_fd(y, x, h, order=2):
    """
    Backward finite difference method for a given order.
    
    Args:
        y (list or array): Array of values.
        x (int or float): Point around which to compute the derivative.
        h (float): Step size.
        order (int): Order of the derivative (1, 2, 3).
        
    Returns:
        float: Approximation of the derivative at point x.
    """
    n = len(y)
    if n < order + 2 * 2:
        raise ValueError("Not enough points to compute the derivative.")
    
    if order == 1:
        return np.round((y[x] - y[x-h]) / h, 6)
    elif order == 2:
        if len(y) >= 3:
            return np.round(((y[x-2*h] - 2*y[x-h] + y[x])/(h**2)), 6)
        else:
            raise ValueError("Not enough points to compute second derivative.")
    elif order == 3:
        if len(y) >= 4:
            return np.round(((-y[x-3*h] + 3*y[x-2*h] - 3*y[x-h] + y[x])/(h**3) * (h/2)), 6)
        else:
            raise ValueError("Not enough points to compute third derivative.")
    else:
        raise ValueError("Order must be 1, 2, or 3.")

def example():
    y = [0.0, 1.0, 2.0, 3.0, 4.0]
    x = 2
    h = 1
    
    # Compute first derivative at x=2 using central method
    df_central_1 = central_fd(y, x, h)
    
    # Compute second derivative at x=2 using forward method
    df_forward_2 = forward_fd(y, x, h, order=2)
    
    print(f"Central FD for first derivative: {df_central_1}")
    print(f"Forward FD for second derivative: {df_forward_2}")

if __name__ == "__main__":
    example()
```
